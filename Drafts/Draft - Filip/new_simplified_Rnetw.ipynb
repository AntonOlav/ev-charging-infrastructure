<h2>Now we have the nodes and edges, we need to calculate the lenght on the edges between the new node-neighbours<h2>

```python
# Set Coordinate System

new_edges_gdf.crs = {'init': 'epsg:4326'}
# Convert the CRS to a projected CRS that uses meters or kilometers as the units
new_edges_gdf = new_edges_gdf.to_crs({'init': 'epsg:4326'}) # Change the EPSG code to the appropriate UTM zone for your data
```

```python
# Using the Haversine formula to calculate length

from math import radians, sin, cos, sqrt, atan2

# define the radius of the Earth in meters
R = 6371000

# create an empty list to store the edge lengths
edge_lengths = []

# loop over the geometry column of the edges_gdf GeoDataFrame
for line in new_edges_gdf['geometry']:
    coords = line.coords
    
    src_lon, src_lat = radians(coords[0][0]), radians(coords[0][1])
    tgt_lon, tgt_lat = radians(coords[1][0]), radians(coords[1][1])
    
    d_lon = tgt_lon - src_lon
    d_lat = tgt_lat - src_lat
    
    a = sin(d_lat/2)**2 + cos(src_lat) * cos(tgt_lat) * sin(d_lon/2)**2
    c = 2 * atan2(sqrt(a), sqrt(1-a))
    distance = R * c
    
    edge_lengths.append(distance)

# add the edge lengths to the edges_gdf GeoDataFrame
new_edges_gdf['length_m'] = edge_lengths

# Convert the length from meters to kilometers
new_edges_gdf['length_km'] = new_edges_gdf['length_m'] / 1000

```

```python
result_df['speedlim'] = 0                                                                       # Create speedlim-column = 0 

for index, row in result_df.iterrows():
    
    source = row['u']                                                                           # get source node-ID
    target = row['v']                                                                           # get target node-ID

    # Store source and target values
    m_source = edge_df.loc[edge_df['source'] == source]                                         #
    m_target = edge_df.loc[edge_df['source'] == target]                                         #

    # if source and target is only one row:
    if len(m_source) == 1 and len(m_target) == 1:
        m_source_speedlim = edge_df.loc[edge_df['source'] == source]['speedlim']                # if one entity, store SOURCE value in one variable
        m_target_speedlim = edge_df.loc[edge_df['source'] == target]['speedlim']                # if one entity, store TARGET value in one variable

        # If source and target are one row AND not NaN
        if not m_source_speedlim.empty and not m_target_speedlim.empty:                         # if source speed limit NOT empty
            m_source_speedlim = m_source_speedlim.values[0]                                     # assign correct source speed limit
            m_target_speedlim = m_target_speedlim.values[0]                                     # assign correct target speed limit

            if m_source_speedlim == m_target_speedlim:                                          # if source and target speed limit is one value AND the same value
                speedlim = m_source_speedlim                                                    # store correct speed limit value 
            else:
                speedlim = (m_source_speedlim + m_target_speedlim)/2                            # store the average speed limit
            result_df.at[index, 'speedlim'] = speedlim                                          # append speedlimit

        # If source and target are one row AND source is not NaN, but target is NaN
        elif not m_source_speedlim.empty and m_target_speedlim.empty:
            speedlim = m_source_speedlim                                                        # store source speed limit
            result_df.at[index, 'speedlim'] = speedlim                                          # append speedlimit
        # If source and target are one row AND source ist NaN, but target is not NaN
        elif  m_source_speedlim.empty and not m_target_speedlim.empty:
            speedlim = m_target                                                                 # store target speed limit
            result_df.at[index, 'speedlim'] = speedlim                                          # append speedlimit

        # Else both source anf target is NaN. Store speed limit = 0 
        else:
            speedlim = 0
            result_df.at[index, 'speedlim'] = speedlim                                          # append speedlimit

    # if source and target has multiple rows:
    elif len(m_source) > 1 and len(m_target) > 1:
        m_source_speedlim = edge_df.loc[edge_df['source'] == source]['speedlim'].tolist()       # if more than one row, make the SOURCE entities as a list
        m_target_speedlim = edge_df.loc[edge_df['source'] == target]['speedlim'].tolist()       # if more than one row, make the TARGET entities as a list

        # if the speed limit in source and target in all rows is the same:
        if sorted(m_source_speedlim) == sorted(m_target_speedlim):
            speedlim = sum(m_source_speedlim)/len(m_source_speedlim)                            # store the average of the speed limits
            result_df.at[index, 'speedlim'] = speedlim                                          # append speedlimit

        # if the speed limit in source and target have one or more common value(s):
        elif any(x in m_source_speedlim for x in m_target_speedlim):
            common_speedlim = list(set(m_source_speedlim).intersection(set(m_target_speedlim)))
            if len(common_speedlim) == 1:
                speedlim = common_speedlim[0]                                                   # store the common speed limit
            else:
                speedlim = sum(common_speedlim)/len(common_speedlim)                            # store the average of the common speed limits
            result_df.at[index, 'speedlim'] = speedlim                                          # append speedlimit
        # else if there are NO common value at all - ERROR
        else:
            result_df.at[index, 'speedlim'] = "Error 1"                                          # append speedlimit

    # if source has multiple rows and target has one:
    elif len(m_source) > 1 and len(m_target) == 1:
        m_source_speedlim = edge_df.loc[edge_df['source'] == source]['speedlim'].tolist()       # if more than one row, make the SOURCE entities as a list
        m_target_speedlim = edge_df.loc[edge_df['source'] == target]['speedlim']                # if one entity, store TARGET value in one variable
        m_target_speedlim = m_target_speedlim.values[0]                                         # assign correct target speed limit

        # if target speedlimit is in source speedlimit list:
        if m_target_speedlim in m_source_speedlim:
            speedlim = m_target_speedlim 
            result_df.at[index, 'speedlim'] = speedlim                                          # append speedlimit
        
        # if target speedlimit is not in source soeedlimit list
        else:
            m_source_speedlim = (sum(m_source_speedlim))/len(m_source_speedlim)                 # average speed limit in source list
            speedlim = (m_source_speedlim + m_target_speedlim) / 2                              # average between sources speedlimit list and target speed limit. 
            result_df.at[index, 'speedlim'] = speedlim                                          # append speedlimit

    # if source has one row and target has multiple rows:
    elif len(m_source) == 1 and len(m_target) > 1:
        m_source_speedlim = edge_df.loc[edge_df['source'] == source]['speedlim']                # if more than one row, make the SOURCE entities as a list
        m_source_speedlim = m_source_speedlim.values[0]                                         # assign correct target speed limit
        m_target_speedlim = edge_df.loc[edge_df['source'] == target]['speedlim'].tolist()       # if more than one row, make the TARGET entities as a list

        # if source speedlimit is in target speedlimit list:
        if m_source_speedlim in m_target_speedlim:
            speedlim = m_source_speedlim 
            result_df.at[index, 'speedlim'] = speedlim                                          # append speedlimit
        
        # if source speedlimit is not in target soeedlimit list
        else:
            m_target_speedlim = (sum(m_target_speedlim))/len(m_target_speedlim)                 # average speed limit in source list
            speedlim = (m_target_speedlim + m_source_speedlim) / 2                              # average between sources speedlimit list and target speed limit. 
            result_df.at[index, 'speedlim'] = speedlim                                          # append speedlimit
    
    else:
        result_df.at[index, 'speedlim'] = "Error 2"                                             # append speedlimit

        
################################################################
# Above is good 
################################################################
```
