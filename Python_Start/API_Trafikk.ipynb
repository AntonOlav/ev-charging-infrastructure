USe the dictonairy coordinates to plot the traffic registration points into a OSMNx map

```python
import osmnx as ox
import networkx as nx

# Load the JSON data from the file
with open("registration_point_data_EandR.json", "r") as f:
    data = json.load(f)

# Create a graph
G = nx.Graph()
G.graph["crs"] = "epsg:4326"  # set the CRS to WGS 84 (latitude/longitude)

# Add nodes to the graph at the coordinates
for point in data["data"]["trafficRegistrationPoints"]:
    coordinates = (point["location"]["coordinates"]["latLon"]["lon"], point["location"]["coordinates"]["latLon"]["lat"])
    G.add_node(point["id"], pos=coordinates)

# Create the map
map_fig = ox.plot_graph(G)

# Display the map
ox.show_figure(map_fig)

```

```python
import osmnx as ox

# Load the JSON data from the file
with open("registration_point_data_EandR.json", "r") as f:
    data = json.load(f)

# Create a list of coordinates
coordinates = [(point["location"]["coordinates"]["latLon"]["lon"], point["location"]["coordinates"]["latLon"]["lat"]) for point in data["data"]["trafficRegistrationPoints"]]

# Set the distance and infrastructure
distance = 10000  # 10 km
infrastructure = "primary|motorway"  # primary roads and motorways

# Generate the graph of roads and highways
G = ox.graph_from_point(coordinates, dist=distance, custom_filter=infrastructure, network_type="drive")

# Create the map
map_fig, ax = ox.plot_graph(G)

# Display the map
ox.show_figure(map_fig)
```
